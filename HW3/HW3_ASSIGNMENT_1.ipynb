```python


mysize = [random.randint(1,2000) for i in range (100)]
mymean=100

print (mysize)
len(mysize)
```

```python
md={'normal':{},'poisson':{},'binomial':{},'chisq':{},'geometric':{},'exponential':{}}
```

```python
for n in mysize:
    md['normal'][n] = np.random.normal(mymean,50, n)
    md['poisson'][n] = np.random.poisson(mymean, n)
    md['binomial'][n] = np.random.binomial(500, 0.2, n)
    md['chisq'][n] = np.random.chisquare(mymean, n)
    md['geometric'][n] = np.random.geometric(0.01, n)
    md['exponential'][n] = np.random.exponential(mymean, n)
```

```python
means={}
means['normal'] = {}
means['poisson'] = {}
means['binomial'] = {}
means['chisq'] = {}
means['geometric'] = {}
means['exponential'] = {}
```

```python
fig_mu_N = pl.figure(figsize=(10,7))
axnormal_mu_n = fig_mu_N.add_subplot(111)
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    plt.title('Normal Distribution') 
    axnormal_mu_n.plot(nn, means['normal'][nn], 'o')
    axnormal_mu_n.set_xlabel('Sample Size')
    axnormal_mu_n.set_ylabel('Mean')
    axnormal_mu_n.plot([min(mysize), max(mysize)],[100, 100], 'k')
    axnormal_mu_n.grid()


       
```

```python
fig_mu_N = pl.figure(figsize=(10,7))
axnormal_mu_n = fig_mu_N.add_subplot(111)
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    plt.title('Poisson Distribution') 
    axnormal_mu_n.plot(nn, means['poisson'][nn], 'o')
    axnormal_mu_n.set_xlabel('Sample Size')
    axnormal_mu_n.set_ylabel('Mean')
    axnormal_mu_n.plot([min(mysize), max(mysize)],[mymean, mymean], 'k')
    axnormal_mu_n.grid(0.1)
```
